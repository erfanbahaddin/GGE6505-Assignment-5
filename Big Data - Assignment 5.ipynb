```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler


url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header=None)
headers = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors",
           "body-style", "drive-wheels", "engine-location", "wheel-base", "length", "width",
           "height", "curb-weight", "engine-type", "num-of-cylinders", "engine-size", "fuel-system",
           "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"]
df.columns = headers

# Replace the missing values with the column means
df = df.replace('?', pd.np.nan) # replace '?' with NaN
df = df.apply(lambda x: x.fillna(x.mean()) if x.dtype != object else x) # replace NaN with mean for numerical columns

# Get the column names of numerical data
numerical_cols = df.select_dtypes(include=[pd.np.number]).columns.tolist()

# Apply min-max normalization to the numerical columns
scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

print(df[numerical_cols].head())
```

```python
import pandas as pd

# Load the data into a pandas dataframe
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header=None)

# Add column names to the dataframe
cols = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors",
        "body-style", "drive-wheels", "engine-location", "wheel-base", "length", "width", "height",
        "curb-weight", "engine-type", "num-of-cylinders", "engine-size", "fuel-system",
        "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm", "city-mpg",
        "highway-mpg", "price"]
df.columns = cols

# Select the numerical columns
numerical_cols = ["symboling", "normalized-losses", "wheel-base", "length", "width", "height",
                  "curb-weight", "engine-size", "bore", "stroke", "compression-ratio",
                  "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"]

# Convert columns to numeric type
df[numerical_cols] = df[numerical_cols].apply(pd.to_numeric, errors="coerce")

# Compute the correlation matrix
corr_matrix = df.corr()

# Print the correlation matrix
print(corr_matrix)

# Visualize the correlation matrix using a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.show()

# Select the most important features based on the correlation matrix
important_cols = ["curb-weight", "engine-size", "horsepower", "highway-mpg", "price"]
df = df[important_cols]

# Print the first few rows of the new dataframe
print(df.head())
```
